```
import requests
from bs4 import BeautifulSoup
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
import pandas as pd
import csv
import validators #checks if urls are valid

def format_text(url): 

    #Implement Requests and BeautifulSoup
    article = requests.get(url)
    #print(article.text)
    article_soup = BeautifulSoup(article.content, "html.parser")

    #Extract Paragraph Text
    article_paragraphs = ''
    for paragraph in article_soup.find_all('p'):
        article_paragraphs += paragraph.text
    #print(article_paragraphs)

    #Tokenize
    tokenizer = nltk.RegexpTokenizer(r"[^\d\W]+[\w]")
    article_tokens = tokenizer.tokenize(article_paragraphs)
    article_tokens = [token.lower() for token in article_tokens]
    #article_tokens

    #Remove Stop Words
    #print(stopwords.words('english'))
    #print(f"Len before stopwords removed: {len(article_tokens)}")
    article_tokens_no_sw = [token for token in article_tokens if not token in stopwords.words('english')]
    #print(f"Len after stopwords removed: {len(article_tokens_no_sw)}")

    #Lemmatize
    lemmatizer = nltk.WordNetLemmatizer()
    article_tokens_no_sw_lem = [lemmatizer.lemmatize(token) for token in article_tokens_no_sw]
    #article_tokens_no_sw_lem

    #Remove Duplicates
    article_tokens_no_sw_lem_unique = list(dict.fromkeys(article_tokens_no_sw_lem))
    #print(article_tokens_no_sw_lem_unique)

    #Sort in Alphabetical Order
    article_tokens_no_sw_lem_unique.sort()
    return article_tokens_no_sw_lem_unique

def retrieve_article(url):
    pass
```

Sweetgreen Press Page

```
URL = "https://www.sweetgreen.com/press"
sg_press = requests.get(URL)
sg_soup = BeautifulSoup(sg_press.content, "html.parser")
#print(sg_soup.prettify())

sg_all_titles = []
titles = sg_soup.find_all("div", class_="link-card__text")
for title in titles:
    sg_all_titles += [title.text.strip()]

sg_header = []
sg_articles_links = []
sg_titles = []

for i in [2,3,4,6]:
    sg_header += ['Sweetgreen']
    sg_articles_links += [sg_soup.find(title=titles[i].text.strip()).get('href').strip()]
    sg_titles += [sg_all_titles[i]]

sg_text = [] 
for article_link in sg_articles_links:
    sg_text += [format_text(article_link)]

sg_zip = zip(sg_header, sg_titles, sg_articles_links, sg_text)
```

```
header = ['Restaurant', 'Article Title', 'Article Link', 'Text']
with open('G2_articles.csv', 'w') as f:
    write = csv.writer(f)
    write.writerow(header)
    write.writerows(sg_zip)
```

Roti Mediterranean QSR Magazine Articles

```
roti_qsr = requests.get("https://www.qsrmagazine.com/chains/roti-modern-mediterranean")
roti_soup = BeautifulSoup(roti_qsr.content, "html.parser")
#print(roti_soup.prettify())

roti_titles_top = roti_soup.find_all("span", class_="field-content top-four-title")
roti_titles_rest = roti_soup.find_all("span", class_="field field--name-title field--type-string field--label-hidden")
roti_top_titles = []
roti_rest_titles = []
for title in roti_titles_top:
    roti_top_titles += [title.text.strip()]
for title in roti_titles_rest:
    roti_rest_titles += [title.text.strip()]

roti_header = []
roti_articles_links = []
roti_titles = []

for i in range(len(roti_titles_top)):
    roti_header += ['Roti']
    roti_articles_links += ['https://www.qsrmagazine.com' + roti_soup.find("a", string=roti_titles_top[i].text.strip()).get('href').strip()]
    roti_titles += [roti_top_titles[i]]
for i in [0,1,3,4,6,7,8]:
    roti_header += ['Roti']
    roti_articles_links += ['https://www.qsrmagazine.com' + roti_soup.find("span", string=roti_titles_rest[i].text.strip()).parent.get('href').strip()]
    roti_titles += [roti_rest_titles[i]]
roti_articles_links

roti_text = [] 
for article_link in roti_articles_links:
    roti_text += [format_text(article_link)]

roti_zip = zip(roti_header, roti_titles, roti_articles_links, roti_text)
```

```
with open('G2_articles.csv', 'a') as f:
    write = csv.writer(f)
    write.writerows(roti_zip)
```

Au Bon Pain Press Page

```
abp_press = requests.get("https://www.aubonpain.com/about/press")
abp_soup = BeautifulSoup(abp_press.content, "html.parser")
#print(abp_soup.prettify())

abp_links = []
abp_titles = []
abp_all_links = abp_soup.find_all("a")
for link in abp_all_links:
    href = link.get("href")
    valid = validators.url(href)
    if valid and "http" in href:
        if not("aubonpain" in href):
            abp_links.append(href)
            abp_titles.append(link.text.strip())

abp_header = []
abp_text = [] 
for article_link in abp_links:
    abp_text += [format_text(article_link)]
    abp_header += ['Au Bon Pain']

abp_zip = zip(abp_header, abp_titles, abp_links, abp_text)
```

```
with open('G2_articles.csv', 'a') as f:
    write = csv.writer(f)
    write.writerows(abp_zip)
```
